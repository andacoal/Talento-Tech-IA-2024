###**Usando funciones**

````python
# Función para clasificar en 3 categorías
def categorias(y):
    if y < 100:
        return "Bajo"
    elif y >= 100 and y <= 140:
        return "Normal"
    else:
        return "Alto"

# Aplicar la función y crear una nueva columna
df['Categorías'] = df['Nivel de Glucosa (mg/dL)'].apply(categorias)
print("\nClasificación en 3 categorías:\n")
df
````

````python
# Función para clasificar en más de 3 categorías
def categorias2(y):
    if y < 80:
        return "Bajo - Muy Bajo"
    elif y >= 80 and y < 100:
        return "Bajo - Moderado"
    elif y >= 100 and y <= 140:
        return "Normal"
    elif y > 140 and y <= 150:
        return "Alto - Moderado"
    else:
        return "Alto - Muy Alto"

# Aplicar la función y crear una nueva columna 'Categoría Más de 3'
df['Categoría_superior a 3'] = df['Nivel de Glucosa (mg/dL)'].apply(categorias2)

print("\nClasificación en más de 3 categorías:\n")
df
````

2. **Desviación estándar de cada variable:**
