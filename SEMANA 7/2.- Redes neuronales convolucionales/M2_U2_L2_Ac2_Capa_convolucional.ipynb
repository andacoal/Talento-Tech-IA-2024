####**RESUMEN**
Una capa convolucional en una red neuronal está diseñada para procesar datos de forma local (ejemplo: uso de imágenes y diferentes factores como clasificación o segmentación).

En este tipo de capas no se usa una conexión completa entre las neuronas de entrada y salida, sino que se aplica un conjunto de filtros —o kernels— de convolución a ciertas regiones locales de la *entrada*. Dichos filtros son matrices bidimensionales que se deslizan sobre la entrada para producir una salida, la cual puede ser interpretada como una representación más abstracta de la entrada original. Cada filtro es como un detector de características, extrayendo estas características de la entrada.

####**Propiedades**


1.   Invariancia a la traslación: estas capas puede detectar patrones independientemente de su ubicación en la imagen, por lo que esto las hace invariantes a cualquier traslación.

2.  Conexión local: cada capa está conectada a una región local de la entrada, reduciendo así el número de parámetros.

3. Compartición de pesos: los filtros se aplican a todas las regiones de la entrada, permitiendo que se puedan compartir los pesos y aprender las características útiles en distintas partes de la imagen.



## Ejemplo de cómo realizar la convolución de una imagen utilizando OpenCV en Python:

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import the alternative function for displaying images

# Cargar la imagen
image = cv2.imread('lobo.jpeg')

# Definir el kernel de convolución
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# Aplicar la convolución a la imagen
convolved_image = cv2.filter2D(image, -1, kernel)

# Mostrar la imagen original y la imagen convolucionada
cv2_imshow(image) # Use cv2_imshow instead of cv2.imshow
cv2_imshow(convolved_image)

# Espera a que se presione una tecla para salir
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Probar con otros filtros

# Definir el kernel de convolución
kernel1 = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]]) # Detección de bordes vertical

kernel2 = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]]) # Detección de bordes horizontal

# Aplicar la convolución a la imagen
convolved_image1 = cv2.filter2D(image, -1, kernel1)
convolved_image2 = cv2.filter2D(image, -1, kernel2)

# Mostrar la imagen original y la imagen convolucionada
print('Imagen original:')
cv2_imshow(image) # Use cv2_imshow instead of cv2.imshow
print('Imagen con filtro 1:')
cv2_imshow(convolved_image1)
print('Imagen con filtro 2:')
cv2_imshow(convolved_image2)

# Espera a que se presione una tecla para salir
cv2.waitKey(0)
cv2.destroyAllWindows()
```
