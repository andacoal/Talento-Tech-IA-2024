La visión computacional es una rama de la inteligencia artificial que se enfoca en la interpretación automática de imágenes y videos por parte de los ordenadores.
Algunas de las aplicaciones más comunes en visión computacional incluyen la clasificación de objetos, la detección de objetos y la segmentación de imágenes.

El problema de clasificación de objetos en imágenes es un campo de la visión computacional que se centra en asignar una etiqueta o categoría a una imagen que contiene uno o varios objetos específicos.

**Etapas del proceso de clasificación de objetos**

> **Preprocesamiento de datos:**
Las imágenes de entrenamiento se preprocesan para garantizar la consistencia en términos de tamaño, formato y calidad. Esto puede incluir operaciones como la normalización de píxeles, el redimensionamiento de imágenes y la corrección de la
iluminación.

> **Extracción de características:**
Se extraen características relevantes de las imágenes que ayudan al modelo a distinguir entre diferentes clases de objetos. Estas características pueden incluir bordes, texturas, colores y formas.

> **Entrenamiento del modelo:**
Se entrena un modelo de clasificación utilizando el conjunto de datos etiquetados.
Durante el entrenamiento, el modelo ajusta sus parámetros para minimizar la diferencia entre las etiquetas predichas y las etiquetas reales de las imágenes de entrenamiento.

> **Evaluación del modelo:**
El modelo entrenado se evalúa utilizando un conjunto de datos de prueba independiente para determinar su rendimiento en la clasificación de objetos en imágenes nuevas y no vistas (se usan métricas como precisión, exhaustividad y puntuación F1).

**Componentes de la detección de objetos**

> **Identificación de objetos:** El primer paso en la detección de objetos es identificar las clases de objetos presentes en la imagen.

> **Localización de objetos:**
Se determina la ubicación espacial precisa de cada objeto dentro de la imagen. Se logra definiendo cajas delimitadoras (bounding boxes) que rodean a cada objeto detectado, especificando sus coordenadas en términos de píxeles o proporciones relativas en la imagen.

> **Clasificación y etiquetado:**
Es necesario clasificarlos correctamente
asignando una etiqueta o categoría a cada uno.


## Ejercicio de Reconocimiento de Rostros de gatos

```python
# Importar bibliotecas
import cv2
import numpy as np
```

```python
# Carga del clasificador de cascada entrenado para detectar rostros de gatos

faceClassif = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
```

```python
# Carga de la imagen de los gatos
image = cv2.imread('gatos.jpg')
image1 = cv2.imread('gatos.jpeg')

# Convertir la imagen a escala de grises
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

Se especifican diferentes parámetros: 'scaleFactor' que especifica cuánto se
reduce la imagen en cada escala de la pirámide de imagen (entre más pequeño el
valor, más precisa la detección pero más tiempo de cálculo); 'minNeighbors' que
es el número mínimo de vecinos que deben encontrarse para que se considere un
objeto; 'minSize' que especifica el tamaño mínimo de los objetos a considerar;
'maxSize' que especifica el tamaño máximo de los objetos a considerar. Y los resultados se almacenan en una variable que contiene las coordenadas *(x, y, w, h)* de los cuadrados delimitadores de los rostros detectados.

```python
# Detección de rostros de gatos
faces = faceClassif.detectMultiScale(gray,
                                     scaleFactor = 1.1,
                                     minNeighbors=5,
                                     minSize=(30,30),
                                     maxSize=(200,200))
print(faces)
print(f'Número de rostros detectados: {len(faces)}')

# Otra imagen
faces1 = faceClassif.detectMultiScale(gray1,
                                     scaleFactor = 1.1,
                                     minNeighbors=5,
                                     minSize=(30,30),
                                     maxSize=(200,200))
print(faces1)
print(f'Número de rostros detectados: {len(faces1)}')
```

```python
# Dibujar rectángulos alrededor de los rostros
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

for (x, y, w, h) in faces1:
    cv2.rectangle(image1, (x, y), (x+w, y+h), (0, 255, 0), 2)
```
