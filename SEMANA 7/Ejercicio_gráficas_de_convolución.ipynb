```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las señales de entrada
x = np.array([1, 2, 1])
h = np.array([1, -1, 2])

# Calcular la convolución
y = np.convolve(x, h, mode='full')

# Definir los índices para las señales
n_x = np.arange(len(x))
n_h = np.arange(len(h))
n_y = np.arange(len(y))

# Crear la figura y los ejes
plt.figure(figsize=(12, 6))

# Graficar la señal de entrada x[n]
plt.subplot(3, 1, 1)
plt.stem(n_x, x, use_line_collection=True)
plt.title('Señal de Entrada x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

# Graficar la respuesta impulsional h[n]
plt.subplot(3, 1, 2)
plt.stem(n_h, h, use_line_collection=True)
plt.title('Respuesta Impulsional h[n]')
plt.xlabel('n')
plt.ylabel('h[n]')

# Graficar la señal de salida y[n]
plt.subplot(3, 1, 3)
plt.stem(n_y, y, use_line_collection=True)
plt.title('Señal de Salida y[n] (Convolución)')
plt.xlabel('n')
plt.ylabel('y[n]')

# Mostrar las gráficas
plt.tight_layout()
plt.show()
```

Otro ejemplo:

```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt

# Definir las secuencias de entrada y respuesta impulsional
X = [2, 1, 3]  # Secuencia de entrada X[n]
h = [1, -1, 2] # Respuesta impulsional h[n]

# Calcular la convolución de X y h
Y = np.convolve(X, h)  # Convolución de X y h, resultado en Y[n]

# Definir los índices para graficar las secuencias
X_indices = np.arange(len(X))  # Índices de X[n] (0, 1, 2)
h_indices = np.arange(len(h))  # Índices de h[n] (0, 1, 2)

# Los índices para Y[n] se calculan considerando el rango total de la convolución
Y_indices = np.arange(len(X) + len(h) - 1)  # Índices de Y[n] (0, 1, 2, 3, 4)

# Crear la gráfica
plt.figure(figsize=(12, 8))

# Graficar X[n]
plt.subplot(3, 1, 1)  # Crear una subgráfica en la primera fila
plt.stem(X_indices, X, use_line_collection=True, basefmt=" ", linefmt='b-', markerfmt='bo')
# Graficar la secuencia X[n]
plt.title('Secuencia de Entrada X[n]')
plt.xlabel('n')  # Etiqueta del eje x
plt.ylabel('X[n]')  # Etiqueta del eje y
plt.grid()

# Graficar h[n]
plt.subplot(3, 1, 2)  # Crear una subgráfica en la segunda fila
plt.stem(h_indices, h, use_line_collection=True, basefmt=" ", linefmt='r-', markerfmt='ro')
# Graficar la respuesta impulsional h[n]
plt.title('Respuesta Impulsional h[n]')
plt.xlabel('n')  # Etiqueta del eje x
plt.ylabel('h[n]')  # Etiqueta del eje y
plt.grid()

# Graficar Y[n] = conv(X, h)
plt.subplot(3, 1, 3)  # Crear una subgráfica en la tercera fila
plt.stem(Y_indices, Y, use_line_collection=True, basefmt=" ", linefmt='g-', markerfmt='go')
# Graficar el resultado de la convolución Y[n]
plt.title('Resultado de la Convolución Y[n]')
plt.xlabel('n')  # Etiqueta del eje x
plt.ylabel('Y[n]')  # Etiqueta del eje y
plt.grid()

plt.tight_layout()  # Ajustar el diseño para evitar superposiciones
plt.show()
```
