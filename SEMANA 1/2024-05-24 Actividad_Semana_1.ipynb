En este caso usaremos como base de ejemplo el *practical_example* proporcionado; pero lo adaptaremos a un nuevo tema que es sobre planetas de nuestro sistema solar.

```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Se crea un dataframe, en este caso, que contiene info. sobre planetas

# Creamos listas con la información de los planetas
planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Neptuno']
diametros = [4879, 12104, 12756, 6792, 142984, 120536, 51118, 49528] # en kilómetros
distancias_al_sol = [57.9, 108.2, 149.6, 227.9, 778.3, 1427, 2871, 4497.1] # en millones de kilómetros

# Creamos un diccionario con la información de los planetas
data_dict = {
    'Planeta': planetas,
    'Diámetro (km)': diametros,
    'Distancia al Sol (millones de km)': distancias_al_sol
}

# Creamos un DataFrame a partir del diccionario
df = pd.DataFrame(data_dict)

# Mostramos el DataFrame
print(df)
```

```python
# Llamamos a la variable
df
```

```python
# prompt: Con DataFrame df: se puede añadir otra columna

# Creamos una lista con el número de lunas de cada planeta, por ejemplo
num_lunas = [0, 0, 1, 2, 79, 82, 27, 14]

# Añadimos la nueva columna al DataFrame
df['Número de lunas'] = num_lunas

# Mostramos el DataFrame
df
```

```python
# Importamos la biblioteca necesaria
import matplotlib.pyplot as plt

# Creamos una gráfica de barras de los diámetros de los planetas
plt.bar(df['Planeta'], df['Diámetro (km)'])

# Añadimos títulos y etiquetas
plt.title('Diámetro de los Planetas')
plt.xlabel('Planeta')
plt.ylabel('Diámetro (km)')

# Mostramos la gráfica
plt.show()
```

### .dtype y .astype

El primer atributo permite obtener el tipo de datos de cierta columna; mientra que el segundo permite cambiar dicho tiempo.

```python
# Obtenemos el tipo de datos de la columna 'Diámetro (km)'
print(df['Diámetro (km)'].dtype)

# Cambiamos el tipo de datos de la columna 'Diámetro (km)' a float
df['Diámetro (km)'] = df['Diámetro (km)'].astype(float)

# Verificamos el cambio
print(df['Diámetro (km)'].dtype)
```
